```python
gc = pd.read_csv('Geospatial_Coordinates (1).csv')
```

```python
gc.set_index("PostalCode", inplace=True)
```

```python
gc1 = gc.loc[['M5G', 'M2H'], ['Latitude', 'Longitude']]
gc2 = gc.loc[['M4B', 'M1J'], ['Latitude', 'Longitude']]
gc3 = gc.loc[['M4G', 'M4M'], ['Latitude', 'Longitude']]
gc4 = gc.loc[['M1R', 'M9V'], ['Latitude', 'Longitude']]
gc5 = gc.loc[['M9L', 'M5V'], ['Latitude', 'Longitude']]
gc6 = gc.loc[['M1B', 'M5A'], ['Latitude', 'Longitude']]
```

```python
initial_stack = pd.concat([gc1, gc2, gc3, gc4, gc5, gc6], axis=0)
initial_stack.reset_index(drop=True, inplace=True)
```

```python
new_stack = pd.concat([new_df, initial_stack], axis=1)
```
